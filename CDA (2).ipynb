```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import glm
import statsmodels.api as sm
```

```python
# Load the dataset
df = pd.read_excel('Digital Mirror Responses Cleaned.xlsx')
```

```python
response = 'D1'
predictors = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8',]
```

```python
# Prepare the data for GLM
X = df[predictors]
X = sm.add_constant(X)  # Add constant term for intercept
y = df[response]
```

```python
# Fit the GLM model
model = glm(f"{response} ~ {' + '.join(predictors)}", data=df, family=sm.families.Binomial()).fit()

# Display the summary
print(model.summary())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_excel('abc.xlsx', sheet_name='Digital Mirror Responses - Form')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import glm

response = 'D1'
predictors = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8']

# Prepare the data for GLM
X = df[predictors]
X = sm.add_constant(X)  # Add constant term for intercept
y = df[response].map({'YES': 1, 'NO': 0})  # Encode response variable

# Fit the GLM model
model = glm(f"{response} ~ {' + '.join(predictors)}", data=df, family=sm.families.Binomial()).fit()

# Display the summary
print(model.summary())
```

```python
import seaborn as sns

# Calculate residuals and fitted values
residuals = model.resid_deviance
fitted_values = model.fittedvalues

plt.figure(figsize=(8, 6))
sns.scatterplot(x=fitted_values, y=residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Deviance Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming Q3 (pressure) and D1 (negative self-esteem effect) are categorical
# Compute the conditional probability of D1 given Q3
contingency_table = pd.crosstab(df['Q3'], df['D1'])

# Convert counts to probabilities
conditional_prob_matrix = contingency_table.div(contingency_table.sum(axis=1), axis=0)

plt.figure(figsize=(10, 6))
sns.heatmap(conditional_prob_matrix, annot=True, cmap='viridis', fmt='.2f', linewidths=0.5)
plt.title('Conditional Probability Matrix: P(D1 | Q3)')
plt.xlabel('D1 (Negative Self-Esteem Effect)')
plt.ylabel('Q3 (Pressure to Conform)')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel('abc.xlsx', sheet_name='Digital Mirror Responses - Form')

# Define the response and additional predictors
response = 'D1'
predictors = ['Q4', 'Q9', 'Q10']  # Including additional categorical variables

# Function to compute conditional probabilities
def compute_conditional_probabilities(df, response, predictor):
    # Create a contingency table
    contingency_table = pd.crosstab(df[predictor], df[response])
    # Convert counts to probabilities
    conditional_prob_matrix = contingency_table.div(contingency_table.sum(axis=1), axis=0)
    return conditional_prob_matrix

# Plotting conditional probability matrices for each predictor
for predictor in predictors:
    prob_matrix = compute_conditional_probabilities(df, response, predictor)
    
    plt.figure(figsize=(10, 6))
    sns.heatmap(prob_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title(f'Conditional Probability Matrix: P({response} | {predictor})')
    plt.xlabel(response)
    plt.ylabel(predictor)
    plt.show()
```
